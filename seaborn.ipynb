PAIRGRID

```python
iris = sns.load_dataset("iris")
iris_g = sns.PairGrid(iris,hue="species")
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
```

```python
iris = sns.load_dataset("iris")
iris_g = sns.PairGrid(iris,hue="species")
iris_g.map(plt.scatter)
iris_g.map_offdiag(plt.scatter)
```

```python
iris = sns.load_dataset("iris")
iris_g = sns.PairGrid(iris,hue="species")
iris_g.map(plt.scatter)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

```python
iris = sns.load_dataset("iris")
iris_g = sns.PairGrid(iris,hue="species",
                      x_vars=["sepal_length","sepal_width"],
                      y_vars=["petal_length","petal_width"])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

FACETGRID

```python
tips_fg = sns.FacetGrid(tips_df,col="time",row="smoker",height=4,aspect=1)
tips_fg.map(plt.hist,"total_bill",bins=8)
```

```python
tips_fg = sns.FacetGrid(tips_df,col="time",row="smoker",height=4,aspect=1.3,
                        col_order=["Lunch","Dinner"],palette="set1")
tips_fg.map(plt.scatter,"total_bill","tip",edgecolor="w")
```

```python
kws = dict(s=50,linewidth=.5,edgecolor = "w")
tips_fg = sns.FacetGrid(tips_df,col="sex",hue = "smoker",height=4,aspect=1,
hue_order=["Yes","No"],
hue_kws = dict(marker = ["^","v"]))
tips_fg.map(plt.scatter,"total_bill","tip",**kws)
```

```python
att_def = sns.load_dataset("attention")
att_def = sns.FacetGrid(att_def,col="subject",col_wrap=5,height=1.5)
att_def.map(plt.plot,"solutions","score",marker=".")
```
